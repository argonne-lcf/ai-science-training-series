```python
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('homework.text', sep=" ", header=None, skiprows=63)
df = df[[2, 7, 8]]
df.rename(columns={2:'step', 7:'loss', 8:'acc'}, inplace=True)

df.loss = df.loss.str.replace('0,loss=', '')
df.loss = df.loss.str.replace(',', '')
df.loss = df.loss.astype(float)

df.acc = df.acc.str.replace('acc=', '')
df.acc = df.acc.astype(float)


# uncomment this in case we want to see loss fluctiations better
df = df[7:]
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(df.step, df.loss, 'g-')
ax2.plot(df.step, df.acc, 'b-')

#ax1.set_yscale("log")
ax1.set_xlabel('Step')
ax1.set_ylabel('Loss', color='g')
ax2.set_ylabel('Accuracy', color='b')

plt.show()
```
