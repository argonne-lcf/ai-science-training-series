```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Homework by Juan Acosta for AI science training series at ANL

I modified the .sh script and the .py to run several tests and store the results in the results in results.txt. Here I just plot the the results as heatmaps.



```python
df_data = pd.read_csv('results.txt', header=None, sep=',', names=['threads', 'buffer', 'mean', 'stdev'])
df_data.head()
```

```python
table_mean = df_data.pivot('threads', 'buffer', 'mean')
ax = sns.heatmap(table_mean)
ax.invert_yaxis()
plt.title('mean image rate')
plt.show()
```

```python
table_stdev = df_data.pivot('threads', 'buffer', 'stdev')
ax = sns.heatmap(table_stdev)
ax.invert_yaxis()
plt.title('stdev of image rate')
plt.show()
```
